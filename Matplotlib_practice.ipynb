```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
%matplotlib inline

x1 = np.linspace(0,10,100)

#create a plot figure
fig = plt.figure()

plt.plot(x1,np.sin(x1),'-')
plt.plot(x1,np.cos(x1),'--')
plt.show()
```

```python
plt.gcf()
plt.gca()
```

```python
#create a plot figure
plt.figure()
plt.subplot(2,1,1) #(rows, columns, panel number)
x1 = np.linspace(0,10,100)
plt.plot(x1,np.sin(x1))

plt.show()

plt.subplot(2,1,2)
plt.plot(x1,np.cos(x1))
plt.show()
```

```python
#State Machine Interface

x1 = np.linspace(0,2,100)

plt.plot(x1,x1,label="Linear")
plt.plot(x1,x1**2,label="Quadratic")
plt.plot(x1,x1**3,label = "Cubic")

plt.xlabel("xlabel")
plt.ylabel('ylabel')

plt.legend()

plt.show()
```

```python
#Formatting the style of plot

plt.plot([1,2,3,4,5],[1,4,9,16,20],'ro')
plt.axis([0,6,0,20]) #[xmin,xmax,ymin,ymax] > Specifies viewport of axes
plt.show()
```

```python
#NumpY ARRAys

t = np.arange(0., 5., 0.2)
print(t)

#Red dashes , blue Squares and green triangles

plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

```python
#Object-oriented API

fig, ax = plt.subplots(2)

#call plot() methods on appropriate object
x1 = np.linspace(0,10,100)

ax[0].plot(x1,np.sin(x1),'b-')
ax[1].plot(x1,np.cos(x1),'b-')
plt.show()

x = np.linspace(0,2,100)

fig, a = plt.subplots(3)

a[0].plot(x,x,'b-')
a[1].plot(x,x**2)
a[2].plot(x,x**3)
plt.show()
```

```python
#Objects and reference

fig = plt.figure()

x2 = np.linspace(0,5,10)
y2 = x2**2

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #It is rect = [left,bottom,width,height]

axes.plot(x2,y2,'r')
axes.set_ylabel('y2')
axes.set_xlabel('x2')
axes.set_title('Title');
```

```python
#Figure and Subplots

fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
plt.show()
```

```python
#Specify foth Lists
x3 = np.arange(0.0,6.0,0.01)

plt.plot(x3,[i**2 for i in x3])
plt.show()
```

```python
#Multiline Plots

x4 = np.arange(1.0,5.0,0.1)

plt.plot(x4,[i*1.5 for i in x4],label = "red")
plt.plot(x4,[i*2.5 for i in x4],label = "blue")
plt.plot(x4,[i*2.0 for i in x4],label = "green")
plt.legend()
plt.show()
```
